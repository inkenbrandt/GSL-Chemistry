```python
%matplotlib inline
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from pylab import mpl
from pylab import rcParams
from pandas.tools.plotting import autocorrelation_plot
rcParams['figure.figsize'] = 15, 10
import urllib2
```

```python
from statsmodels.tsa.arima_model import ARIMA

def Plot_ARIMA(co2,dates,p,d,q,freq):
    arima = ARIMA(co2, [p, d, q], exog=None, dates=dates, freq=freq, missing='none')
    arima_results = arima.fit(trend='c', disp=False)
    co2_ARIMA = arima_results.predict(exog=None, dynamic=False)
    
    if d>0:
        co2_d = co2.diff()[d:]
        dates_d = dates[d:]
    else:
        co2_d = co2

    plt.figure(figsize=(20,5))
    plt.plot(dates[d:],co2_d)
    plt.plot(dates[d:],co2_ARIMA)
    return arima_results.aic
```

```python
infile = r'http://waterdata.usgs.gov/nwis/dv?cb_72020=on&format=rdb&site_no=10010000&referred_module=sw&period=&begin_date=2012-01-01&end_date=2015-05-05'
response = urllib2.urlopen(infile)
df = pd.read_table(response, sep='\t', names=['agency_cd','site_no','datetime','elev','status'],index_col=['datetime'], skiprows=28, parse_dates=True)
df.drop(['agency_cd','site_no'],inplace=True,axis=1)
freq = '7D'
df = df.resample(freq)
df = df.interpolate(method='time')
```

```python
plt.figure()
x = df.index
y = df.elev
df['delta']=y.diff()
z = df.delta
#y = df.delta
plt.plot(x,y)
plt.scatter(x,y)
plt.xlim('2010-01-01','2015-01-01')
plt.ylim(4190,4200)
```

```python
index = np.arange(len(y))
dates=x
y_lin = stats.linregress(index, y)
y_predict = y_lin[0] * index + y_lin[1]

#figure(figsize=(20,5))
plt.plot(dates, y)
plt.plot(dates, y_predict)
plt.title('GSL Elev')
plt.xlabel('Year')
plt.ylabel('elev (ft)')
plt.figtext(0.15, 0.25, 'r = '+str(y_lin[2]), size='x-large')
plt.figtext(0.15, 0.2, 'intercept = '+str(y_lin[1]), size='large')
plt.figtext(0.15, 0.15, 'slope = '+str(y_lin[0]), size='large')
print index
```

```python
df['delin']=df['elev']-y_predict
plt.plot(dates,df.delin)
```

```python

from scipy.optimize import curve_fit
data = df['delin']

ix = pd.DatetimeIndex(start=df.index[1], end='2016-01-31', freq=freq)
df1 = df.reindex(ix)

x = df.index.to_julian_date()
guess_freq = .018
guess_amplitude = 5*np.std(y_predict)
guess_phase = 0.01
guess_offset = np.mean(y_predict)

p0=[guess_freq, guess_amplitude,
    guess_phase, guess_offset]

# create the function we want to fit
def my_sin(x, freq, amplitude, phase, offset):
    return np.sin(x * freq + phase) * amplitude + offset

# now do the fit
fit = curve_fit(my_sin, x, data, p0=p0)

x = df1.index.to_julian_date()
# recreate the fitted curve using the optimized parameters
data_fit = my_sin(x, *fit[0])

plt.plot(data, '.')
plt.plot(data_fit, label='after fitting')
trend = []
for i in range(len(data_fit)):
    trend.append(i+y_lin[0]+y_lin[1])

plt.legend()
print(fit)
```
